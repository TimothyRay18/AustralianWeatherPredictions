Kami memutuskan untuk memprediksi apakah akan terjadi hujan pada besok hari di Australia dengan menggunakan regresi logistik karena regresi logistik adalah bentuk khusus regresi yang diformulasikan untuk melakukan klasifikasi data ke dalam dua group (prediksi group) dan menjelaskan variabel dependen biner (kategorikal/non-metric), dalam hal ini variabel dependen adalah "RainTomorrow"

Kami merasa variabel "RainTomorrow" penting karena berpengaruh dalam berbagai bidang seperti industri transportasi udara, darat, dan laut. Tak hanya itu bidang pariwisata dan kehidupan sehari-hari masyarakat pun ikut mendapatkan manfaat dari hasil prediksi ini. Ke depannya, kami berharap hasil analisis kami dapat dikembangkan lebih lanjut untuk memprediksi kemungkinan bencana alam di masa yang akan datang.

## Random Forest with PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

max_depth1, max_depth2, max_depth3 = 4, 10, 20
max_leaf_nodes1, max_leaf_nodes2, max_leaf_nodes3 = 5, 10, 15
hyperparams = (max_depth1, max_leaf_nodes1), (max_depth1, max_leaf_nodes2), (max_depth1, max_leaf_nodes3), (max_depth2, max_leaf_nodes1), (max_depth2, max_leaf_nodes2), (max_depth2, max_leaf_nodes3), (max_depth3, max_leaf_nodes1), (max_depth3, max_leaf_nodes2), (max_depth3, max_leaf_nodes3)

rnd_clfs = []
for max_depth, max_leaf_nodes in hyperparams:
    rnd_pipe_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("rnd_clf", RandomForestClassifier(n_estimators=1000, max_leaf_nodes=max_leaf_nodes, max_depth=max_depth, random_state=42))
        ])
    rnd_pipe_clf.fit(x_train, y_train)
    rnd_clfs.append(rnd_pipe_clf)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import confusion_matrix
hyperparams = [[4, 5], [4, 10], [4, 15], [10, 5], [10, 10], [10, 15], [20, 5], [20, 10], [20, 15]]
for i, rnd_clf in enumerate(rnd_clfs):
    print("Random Forest "+str(i+1))
    predictions = rnd_clf.predict(x_test)
    errors = abs(predictions - y_test)
    print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
    
    accuracy = accuracy_score(predictions, y_test)*100
    accuracy = round(accuracy,1)
    print("Accuracy "+ str(accuracy))
    print("max_depth = "+ str(hyperparams[i][0]) +" | max_leaf_nodes = "+ str(hyperparams[i][1]))
    
    matrix = confusion_matrix(y_test,predictions)
    sns.heatmap(matrix/np.sum(matrix), annot=True, fmt='.2%', cmap='Blues').set(title="Confusion Matrix")
    print(classification_report(y_test,predictions))
    plt.show()
    print()
```

## Random Forest Regressor with PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

max_depth1, max_depth2, max_depth3 = 4, 10, 20
max_leaf_nodes1, max_leaf_nodes2, max_leaf_nodes3 = 5, 10, 15
hyperparams = (max_depth1, max_leaf_nodes1), (max_depth1, max_leaf_nodes2), (max_depth1, max_leaf_nodes3), (max_depth2, max_leaf_nodes1), (max_depth2, max_leaf_nodes2), (max_depth2, max_leaf_nodes3), (max_depth3, max_leaf_nodes1), (max_depth3, max_leaf_nodes2), (max_depth3, max_leaf_nodes3)

rnd_clfs = []
for max_depth, max_leaf_nodes in hyperparams:
    rnd_pipe_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("rnd_clf", RandomForestRegressor(n_estimators=1000, max_leaf_nodes=max_leaf_nodes, max_depth=max_depth, random_state=42))
        ])
    rnd_pipe_clf.fit(x_train, y_train)
    rnd_clfs.append(rnd_pipe_clf)
```

```python
def percentage_error(actual, predicted):
    res = np.empty(actual.shape)
    for j in range(actual.shape[0]):
        if actual[j] != 0:
            res[j] = (actual[j] - predicted[j]) / actual[j]
        else:
            res[j] = predicted[j] / np.mean(actual)
    return res

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs(percentage_error(np.asarray(y_true), np.asarray(y_pred)))) * 100
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import confusion_matrix
hyperparams = [[4, 5], [4, 10], [4, 15], [10, 5], [10, 10], [10, 15], [20, 5], [20, 10], [20, 15]]
for i, rnd_clf in enumerate(rnd_clfs):
    print("Random Forest "+str(i+1))
    predictions = rnd_clf.predict(x_test)
    errors = abs(predictions - y_test)
    print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
    # Calculate mean absolute percentage error (MAPE)
    mape = 100 * (errors / y_test)
    # Calculate and display accuracy
    accuracy = 100 - np.mean(mape)
    print('Accuracy:', round(accuracy, 2), '%.')
    df_perform = pd.DataFrame(predictions, columns = ['Prediction'])
    df_perform['Target'] = y_test.reset_index(drop=True)
    df_perform['Residual'] = df_perform['Target'] - df_perform['Prediction']
    df_perform['Difference%'] = np.absolute(df_perform['Residual']/df_perform['Target']*100)
    print(df_perform.describe())
    print()
```
